# **Join (Merge)**

```python
df.columns
```

```python
df.info()
```

```python
import numpy as np
import pandasql as ps
```

```python
data1 = {
    'ID': [1, 2, 3, 4, 5],
    'PM2.5 (ug/m3)': [12.5, 15.3, np.nan, 18.7, 20.1],
    'PM10 (ug/m3)': [25.1, 30.5, 22.3, np.nan, 31.4],
    'NO (ug/m3)': [5.1, 7.2, np.nan, 6.8, 8.0],
    'NO2 (ug/m3)': [12.4, np.nan, 13.8, 14.1, 15.5]
}

df1 = pd.DataFrame(data1, columns=['ID', 'PM2.5 (ug/m3)', 'PM10 (ug/m3)', 'NO (ug/m3)', 'NO2 (ug/m3)'])
```

```python
data2 = {
    'ID': [3, 4, 5, 6, 7],
    'NOx (ppb)': [50, 60, np.nan, 70, 80],
    'NH3 (ug/m3)': [1.5, 2.0, 2.5, np.nan, 3.0],
    'SO2 (ug/m3)': [0.5, 0.6, 0.7, 0.8, 0.9],
    'Temp (degree C)': [20.1, 22.4, np.nan, 24.0, 23.5]
}

df2 = pd.DataFrame(data2)
```

# Inner Join

```python
inner_join_df = pd.merge(df1, df2, on='ID', how='inner')
inner_join_df
```

```python
Query_string_inner = """SELECT *
                        FROM df1 AS a
                        INNER JOIN df2 AS b
                        ON a.ID = b.ID;"""

inner_join_result = ps.sqldf(Query_string_inner, globals())
inner_join_result
```

# Left Join

```python
left_join_df = pd.merge(df1, df2, on='ID', how='left')
left_join_df
```

```python
Query_string_left = """SELECT *
                       FROM df1 AS a
                       LEFT JOIN df2 AS b
                       ON a.ID = b.ID;"""

ps.sqldf(Query_string_left, globals())
```

# Right Join

```python
right_join_df = pd.merge(df1, df2, on='ID', how='right')
right_join_df
```

# Outer Join

```python
outer_join_df = pd.merge(df1, df2, on='ID', how='outer')
outer_join_df
```

```python
Query_string_outer = """SELECT *
                        FROM df1 AS a
                        LEFT JOIN df2 AS b
                        ON a.ID = b.ID
                        UNION
                        SELECT *
                        FROM df2 AS b
                        LEFT JOIN df1 AS a
                        ON a.ID = b.ID;"""

outer_join_result = ps.sqldf(Query_string_outer, globals())
outer_join_result
```

# **Shifting**

Shifting is also known as “lag” and moves a value back or forward in time.

Shift an index by the desired number of periods with an optional time frequency.

```python
df.shift(periods=3).head(7) #it shifts the data downward (or forward) by the specified number of periods.
```

```python
df.shift(periods=-1).head(7) #it shifts the data upward (or backward) by the specified number of periods.
```

##shifting the data along the columns, rather than the rows. This means that the values in each column will be shifted horizontally rather than vertically.

```python
df.shift(periods=3, axis =1).head(7)
```

```python
df.shift(periods=3,fill_value=0).head(7)
```

```python
df.shift(periods=3,fill_value=0,axis =1).head(7)
```

# BFill

###Backward fill involves filling missing values with the next known value that follows them. This method assumes that the next observed value is the best estimate for the missing values.



```python
df.bfill()
```

# FFill

###Forward fill involves filling missing values with the last known value that precedes them. This method assumes that the last observed value is the best estimate for the missing values.

```python
df.ffill()
```

# FillNA

```python
df.fillna(10)
```
